```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [567,56,78,35,84]
plt.plot(y)
```

```python
plt.figure(figsize=(10,10))
plt.pie(y)
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1,figsize= (15,5))
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3*np.pi , 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(15,2))
ax[0,][0].plot(t, x,"r")
ax[0][1].bar(t,y)
plt.show
```

```python
x = [100,200,300,400,500,600,700]
y = [23,45,34,65,2,2,47]
plt.plot(x,y,linewidth=2,color='purple',marker='o',markersize=15,markerfacecolor='white',markeredgecolor='red',linestyle='--')
```

```python
x = [100,200,300,400,500,600,700]
y = [23,45,34,65,2,2,47]
plt.plot(x,y, 'go--', linewidth=2, markersize=10, markerfacecolor = 'white', markeredgecolor = 'red' )
```
