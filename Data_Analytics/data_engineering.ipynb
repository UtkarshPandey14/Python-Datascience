# column related operation

```python
df.rename(mapper={"make":"brand"},axis=1, inplace = True)
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'],inplace = True)
```

```python
df.head()
```

```python
df.replace('alfa-romero','alfa-romeo',inplace=True)
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['price'],inplace= True)
```

```python
df.info()
```

```python
df.isnull().sum()     #count the no.of missing data in each column
```

```python
mnl = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan,mnl, inplace= True)
df.head()
```

```python
mnll = df['bore'].astype('float').mean()
```

```python
df['bore'].replace(np.nan,mnll, inplace= True)
df.head()
```

```python
mnll = df['stroke'].astype('float').mean()
```

```python
df['stroke'].replace(np.nan,mnll, inplace= True)
df.head()
```

```python
mnll = df['horsepower'].astype('float').mean()
```

```python
df['horsepower'].replace(np.nan,mnll, inplace= True)
df.head()
```

```python
mnll = df['peak-rpm'].astype('float').mean()
```

```python
df['peak-rpm'].replace(np.nan,mnll, inplace= True)
df.head()
```

```python
df.isnull().sum() 
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("automobile_cleaned.csv",index=False)
```
